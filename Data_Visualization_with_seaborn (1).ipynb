```python
palette = ['#A1C9F4','#D0BBFF','#DEBB9B','#FAB0E4']
sns.set()
sns.catplot(x = "day" , y = "total_bill" , data = tips , palette = palette)
```

```python
sns.displot(data = tips , x = "total_bill" , hue = "sex" , col = "day" , kde = True)
```

```python
sns.histplot(tips["total_bill"] , kde = True)
plt.show()
```

```python
sns.jointplot(x = "total_bill" , y = "tip" , data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill' , y = 'tip' , data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill' , y = 'tip' , data = tips , hue = "size" , palette = "deep" ,
 hue_norm = (0,7) , size = "size" , sizes = (20,200) , legend = "full")
```

```python
sns.pairplot(tips)
plt.show()
```
